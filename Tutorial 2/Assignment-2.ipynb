```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3])
y = np.array([1.2, 1.9, 3.2])
```

```python
plt.plot(x, y, 'ro',alpha=0.8)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Sample Distribution")
```

```python
def max_likelihood(x, y):
    w1 = (len(x) * x.dot(y.T) - (np.sum(y) * np.sum(x))) / (len(x) * x.dot(x.T) - np.sum(x) ** 2)
    w0 = np.mean(y) - w1 * np.mean(x)
    
    return w1, w0
```

```python
w1, w0 = max_likelihood(x, y)
print(w1 , w0)
```

```python
plt.plot(x, w1 * x + w0)
plt.plot(x, y , 'ro')
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Best Fit')
```

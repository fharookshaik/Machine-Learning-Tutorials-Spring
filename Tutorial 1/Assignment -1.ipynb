```python
import numpy as np 
from matplotlib import pyplot as plt
from scipy.stats import norm
```

```python
mean = 5
var = 1
sd = np.sqrt(var)
```

```python
# Question 1

x = np.linspace(0,10,10,False)
y = norm.pdf(x,mean,sd)

plt.plot(x,y,'ro')
plt.title("Gaussian Distribution of 10 Sample points")
```

```python
# Question 2
plt.figure(figsize=(30,60))
# plt.suptitle("Likelihood Function for mean 0 to 10",fontsize=32)
for i in range(11):
    plt.subplot(10,3,i+1)
    z = norm.pdf(x,i,1)
    plt.plot(x,y,'ro')
    plt.plot(x,z,'g--o')
    plt.title("Mean = {0} & Variance = 1".format(i))
```
